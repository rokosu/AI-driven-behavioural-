### ðŸ”¬ MARKET MICROSCOPE: KENYA'S 2024 GREEN BOND ISSUANCE

```python
import matplotlib.pyplot as plt
import numpy as np

# Kenya's 2024 Green Bond Metrics
metrics = {
    'Issue Size': 500,  # $M
    'Yield': 9.75,      # %
    'Oversubscription': 4.5,  # times
    'African Investors': 40,  # %
    'Maturity': 7       # years
}

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 8))

# 1. Yield Comparison
countries = ['Kenya', 'Egypt', 'Ghana', 'SA', 'US Treasury']
yields = [9.75, 8.2, 14.5, 7.3, 4.2]
colors = ['green' if x == 'Kenya' else 'gray' for x in countries]
ax1.bar(countries, yields, color=colors)
ax1.set_title('Green Bond Yield vs. Regional Peers')
ax1.set_ylabel('Yield (%)')

# 2. Investor Distribution
investors = ['European', 'African', 'Asian', 'Other']
distribution = [45, 40, 10, 5]
ax2.pie(distribution, labels=investors, autopct='%1.0f%%')
ax2.set_title('Investor Geographic Mix')

# 3. Use of Proceeds
projects = ['Geothermal', 'Water Mgmt', 'Green Bldgs', 'Agri']
allocations = [60, 25, 10, 5]
ax3.barh(projects, allocations, color='#2E8B57')
ax3.set_title('Proceeds Allocation (%)')

# 4. Timeline
milestones = ['Framework', 'Roadshow', 'Bookbuilding', 'Pricing', 'Listing']
dates = ['Nov 23', 'Jan 24', 'Feb 24', 'Feb 24', 'Mar 24']
ax4.plot(range(5), [1]*5, 'o-', linewidth=3, markersize=10)
ax4.set_xticks(range(5))
ax4.set_xticklabels(milestones, rotation=45)
ax4.set_title('Issuance Timeline')
ax4.set_ylim(0.5, 1.5)

plt.tight_layout()
plt.savefig('kenya_green_bond_case.png', dpi=300)
plt.show()
```

The Virtuous Cycle of Credibility

```python
import networkx as nx

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Virtuous Cycle
G1 = nx.DiGraph()
virtuous_nodes = ['Strong Institutions', 'Policy Consistency', 'Positive Ratings',
                  'Lower Borrowing Costs', 'Growth Investment', 'Enhanced Credibility']
G1.add_edges_from([
    ('Strong Institutions', 'Policy Consistency'),
    ('Policy Consistency', 'Positive Ratings'),
    ('Positive Ratings', 'Lower Borrowing Costs'),
    ('Lower Borrowing Costs', 'Growth Investment'),
    ('Growth Investment', 'Enhanced Credibility'),
    ('Enhanced Credibility', 'Strong Institutions')  # Complete cycle
])

pos1 = nx.circular_layout(G1)
nx.draw_networkx(G1, pos1, with_labels=True, node_color='lightgreen',
                node_size=3000, font_size=9, ax=ax1, arrows=True,
                edge_color='green', width=2)
ax1.set_title('Virtuous Credibility Cycle', fontweight='bold')
ax1.axis('off')

# Vicious Cycle
G2 = nx.DiGraph()
vicious_nodes = ['Weak Governance', 'Policy Volatility', 'Rating Downgrades',
                 'Higher Borrowing Costs', 'Reduced Investment', 'Credibility Erosion']
G2.add_edges_from([
    ('Weak Governance', 'Policy Volatility'),
    ('Policy Volatility', 'Rating Downgrades'),
    ('Rating Downgrades', 'Higher Borrowing Costs'),
    ('Higher Borrowing Costs', 'Reduced Investment'),
    ('Reduced Investment', 'Credibility Erosion'),
    ('Credibility Erosion', 'Weak Governance')  # Complete cycle
])

pos2 = nx.circular_layout(G2)
nx.draw_networkx(G2, pos2, with_labels=True, node_color='lightcoral',
                node_size=3000, font_size=9, ax=ax2, arrows=True,
                edge_color='red', width=2)
ax2.set_title('Vicious Distrust Cycle', fontweight='bold')
ax2.axis('off')

plt.suptitle('How Credibility Cycles Become Self-Reinforcing', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('credibility_cycles.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import matplotlib.patches as mpatches

plt.figure(figsize=(16, 12))

# Create directed graph
G = nx.DiGraph()

# Add nodes with categories
nodes = {
    # Main categories (Central nodes)
    'SOVEREIGN\nCREDIBILITY': {'color': '#1E3A8A', 'size': 4000, 'pos': (0.5, 0.5)},

    # Issuers (Left side)
    'SOVEREIGNS\n(Nations)': {'color': '#3B82F6', 'size': 1800, 'pos': (0.1, 0.7)},
    'CORPORATES': {'color': '#60A5FA', 'size': 1600, 'pos': (0.1, 0.3)},

    # Instruments (Top center)
    'BONDS': {'color': '#10B981', 'size': 1500, 'pos': (0.3, 0.85)},
    'SUKUK': {'color': '#10B981', 'size': 1300, 'pos': (0.4, 0.9)},
    'GREEN BONDS': {'color': '#10B981', 'size': 1300, 'pos': (0.5, 0.85)},
    'SUSTAINABILITY-\nLINKED': {'color': '#10B981', 'size': 1300, 'pos': (0.6, 0.9)},
    'DIASPORA BONDS': {'color': '#10B981', 'size': 1300, 'pos': (0.7, 0.85)},

    # Intermediaries (Right side)
    'RATING\nAGENCIES': {'color': '#F59E0B', 'size': 1600, 'pos': (0.9, 0.8)},
    'INVESTMENT\nBANKS': {'color': '#F59E0B', 'size': 1500, 'pos': (0.9, 0.6)},
    'EXCHANGES': {'color': '#F59E0B', 'size': 1400, 'pos': (0.9, 0.4)},
    'REGULATORS': {'color': '#F59E0B', 'size': 1400, 'pos': (0.9, 0.2)},

    # Investors (Bottom)
    'INSTITUTIONAL\n(Pensions, SWFs)': {'color': '#8B5CF6', 'size': 1700, 'pos': (0.2, 0.1)},
    'RETAIL': {'color': '#8B5CF6', 'size': 1400, 'pos': (0.4, 0.05)},
    'MULTILATERAL\n(Banks, IMF)': {'color': '#8B5CF6', 'size': 1600, 'pos': (0.6, 0.1)},
    'HEDGE\nFUNDS': {'color': '#8B5CF6', 'size': 1400, 'pos': (0.8, 0.05)},

    # Narratives (Top)
    'MEDIA': {'color': '#EF4444', 'size': 1500, 'pos': (0.2, 0.95)},
    'ANALYSTS &\nRESEARCH': {'color': '#EF4444', 'size': 1600, 'pos': (0.5, 0.95)},
    'POLICYMAKERS &\nGOVERNMENTS': {'color': '#EF4444', 'size': 1800, 'pos': (0.8, 0.95)},
}

# Add nodes to graph
for node, attrs in nodes.items():
    G.add_node(node, pos=attrs['pos'], color=attrs['color'], size=attrs['size'])

# Define connections (edges) with different line styles for different flows
edges = [
    # Capital Flow (Blue solid lines)
    ('SOVEREIGNS\n(Nations)', 'SOVEREIGN\nCREDIBILITY', {'color': '#3B82F6', 'width': 3, 'style': 'solid', 'label': 'Creates'}),
    ('SOVEREIGN\nCREDIBILITY', 'BONDS', {'color': '#3B82F6', 'width': 3, 'style': 'solid', 'label': 'Issues'}),
    ('BONDS', 'INSTITUTIONAL\n(Pensions, SWFs)', {'color': '#3B82F6', 'width': 2.5, 'style': 'solid'}),
    ('BONDS', 'INVESTMENT\nBANKS', {'color': '#3B82F6', 'width': 2.5, 'style': 'solid'}),
    ('INVESTMENT\nBANKS', 'RETAIL', {'color': '#3B82F6', 'width': 2, 'style': 'solid'}),

    # Information Flow (Red dashed lines)
    ('RATING\nAGENCIES', 'BONDS', {'color': '#EF4444', 'width': 2, 'style': 'dashed', 'label': 'Rates'}),
    ('ANALYSTS &\nRESEARCH', 'INSTITUTIONAL\n(Pensions, SWFs)', {'color': '#EF4444', 'width': 2, 'style': 'dashed', 'label': 'Advises'}),
    ('MEDIA', 'RETAIL', {'color': '#EF4444', 'width': 2, 'style': 'dashed', 'label': 'Informs'}),
    ('POLICYMAKERS &\nGOVERNMENTS', 'REGULATORS', {'color': '#EF4444', 'width': 2, 'style': 'dashed', 'label': 'Directs'}),

    # Trust/Feedback Flow (Green dotted lines)
    ('INSTITUTIONAL\n(Pensions, SWFs)', 'SOVEREIGN\nCREDIBILITY', {'color': '#10B981', 'width': 2, 'style': 'dotted', 'label': 'Validates'}),
    ('RETAIL', 'SOVEREIGN\nCREDIBILITY', {'color': '#10B981', 'width': 1.5, 'style': 'dotted'}),
    ('MULTILATERAL\n(Banks, IMF)', 'SOVEREIGN\nCREDIBILITY', {'color': '#10B981', 'width': 2, 'style': 'dotted', 'label': 'Accredits'}),
    ('EXCHANGES', 'SOVEREIGN\nCREDIBILITY', {'color': '#10B981', 'width': 1.5, 'style': 'dotted', 'label': 'Lists'}),

    # Regulatory Flow (Orange dash-dot lines)
    ('REGULATORS', 'EXCHANGES', {'color': '#F59E0B', 'width': 2, 'style': 'dashdot', 'label': 'Oversees'}),
    ('REGULATORS', 'INVESTMENT\nBANKS', {'color': '#F59E0B', 'width': 2, 'style': 'dashdot'}),
    ('EXCHANGES', 'BONDS', {'color': '#F59E0B', 'width': 2, 'style': 'dashdot', 'label': 'Trades'}),
]

# Add edges to graph
for edge in edges:
    G.add_edge(edge[0], edge[1], **edge[2])

# Get positions for drawing
pos = nx.get_node_attributes(G, 'pos')
node_colors = [G.nodes[n]['color'] for n in G.nodes()]
node_sizes = [G.nodes[n]['size'] for n in G.nodes()]

# Create figure with better layout
plt.figure(figsize=(18, 14))

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes,
                       alpha=0.95, edgecolors='white', linewidths=3)

# Draw node labels
for node, (x, y) in pos.items():
    # Split label for multiline
    lines = node.split('\n')
    for i, line in enumerate(lines):
        offset = (len(lines)-1)*0.04 - i*0.08
        plt.text(x, y + offset, line,
                fontsize=10 if 'CREDIBILITY' in node else 9,
                fontweight='bold' if 'CREDIBILITY' in node else 'normal',
                ha='center', va='center',
                color='white' if 'CREDIBILITY' in node else 'black')

# Draw edges with different styles
edge_styles = {'solid': [], 'dashed': [], 'dotted': [], 'dashdot': []}
edge_colors = {'solid': [], 'dashed': [], 'dotted': [], 'dashdot': []}
edge_widths = {'solid': [], 'dashed': [], 'dotted': [], 'dashdot': []}

for u, v, data in G.edges(data=True):
    style = data.get('style', 'solid')
    edge_styles[style].append((u, v))
    edge_colors[style].append(data.get('color', '#000000'))
    edge_widths[style].append(data.get('width', 1))

# Draw each edge style separately
for style in edge_styles:
    if edge_styles[style]:
        nx.draw_networkx_edges(G, pos, edgelist=edge_styles[style],
                              edge_color=edge_colors[style],
                              width=edge_widths[style],
                              style=style,
                              alpha=0.7,
                              arrowstyle='-|>', arrowsize=15,
                              connectionstyle='arc3,rad=0.1')

# Add edge labels (only for important flows)
edge_labels = {
    ('SOVEREIGNS\n(Nations)', 'SOVEREIGN\nCREDIBILITY'): 'CREATES',
    ('SOVEREIGN\nCREDIBILITY', 'BONDS'): 'ISSUES',
    ('RATING\nAGENCIES', 'BONDS'): 'RATES',
    ('ANALYSTS &\nRESEARCH', 'INSTITUTIONAL\n(Pensions, SWFs)'): 'ADVISES',
    ('MEDIA', 'RETAIL'): 'INFORMS',
    ('POLICYMAKERS &\nGOVERNMENTS', 'REGULATORS'): 'DIRECTS',
    ('INSTITUTIONAL\n(Pensions, SWFs)', 'SOVEREIGN\nCREDIBILITY'): 'VALIDATES',
    ('EXCHANGES', 'SOVEREIGN\nCREDIBILITY'): 'LISTS',
    ('REGULATORS', 'EXCHANGES'): 'OVERSEES',
    ('EXCHANGES', 'BONDS'): 'TRADES',
}

# Draw edge labels
for (u, v), label in edge_labels.items():
    x = (pos[u][0] + pos[v][0]) / 2
    y = (pos[u][1] + pos[v][1]) / 2

    # Adjust position for better visibility
    dx = pos[v][0] - pos[u][0]
    dy = pos[v][1] - pos[u][1]

    # Perpendicular offset
    offset_x = -dy * 0.05
    offset_y = dx * 0.05

    plt.text(x + offset_x, y + offset_y, label,
            fontsize=8, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

# Add category titles with colored backgrounds
category_titles = [
    ('ISSUERS', (0.05, 0.5), '#3B82F6'),
    ('INSTRUMENTS', (0.5, 0.99), '#10B981'),
    ('INTERMEDIARIES', (0.95, 0.5), '#F59E0B'),
    ('INVESTORS', (0.5, -0.02), '#8B5CF6'),
    ('NARRATIVES', (0.5, 1.08), '#EF4444'),
]

for title, position, color in category_titles:
    plt.text(position[0], position[1], title,
            fontsize=16, fontweight='bold', color='white',
            ha='center', va='center',
            bbox=dict(boxstyle='round,pad=0.8', facecolor=color, edgecolor='white', linewidth=2))

# Add flow type legend
flow_types = [
    ('Capital Flow', '#3B82F6', 'solid'),
    ('Information Flow', '#EF4444', 'dashed'),
    ('Trust/Feedback', '#10B981', 'dotted'),
    ('Regulatory Flow', '#F59E0B', 'dashdot'),
]

legend_x = 0.02
legend_y = 0.98
for i, (label, color, style) in enumerate(flow_types):
    y_pos = legend_y - i * 0.03
    plt.plot([legend_x, legend_x + 0.05], [y_pos, y_pos],
            color=color, linestyle=style, linewidth=3)
    plt.text(legend_x + 0.06, y_pos, label, fontsize=9, va='center')

# Add central concept explanation
plt.text(0.5, 0.4, 'TRUST\nFLOWS', fontsize=20, fontweight='bold',
        color='white', ha='center', va='center',
        bbox=dict(boxstyle='circle,pad=1', facecolor='#1E3A8A', edgecolor='white', linewidth=3))

# Set plot limits and remove axes
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.2)
plt.axis('off')

# Add title
plt.title('Figure 2.1: Global Capital Market Ecosystem Map\nHow Trust Flows Through the Financial System',
         fontsize=16, fontweight='bold', pad=20)

# Save high-resolution image
plt.tight_layout()
plt.savefig('global_capital_ecosystem.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()

print("âœ“ Global Capital Market Ecosystem Map saved as 'global_capital_ecosystem.png'")
```
